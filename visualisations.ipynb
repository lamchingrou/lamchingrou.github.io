```python
import pandas as pd
from tapy import Indicators
from yahoo_fin.stock_info import get_data
```

```python
amzn = get_data("amzn", start_date= "01/01/2021", end_date= "25/08/2021", index_as_date= False, interval="1d")
```

```python
amzn.rename(columns={'open': 'Open', 'high': 'High', 'low': 'Low', 'close': 'Close'}, inplace=True)
```

```python
i = Indicators(amzn)
i.fractals()
amzn_fractals = i.df
```

```python
import numpy as np
```

```python
bullish_markers = []
for index,row in amzn_fractals.iterrows():
    if row['fractals_low'] == True:
        bullish_markers.append(row['Low']*0.9)
    else:
        bullish_markers.append(np.nan)
        
bearish_markers = []
for index,row in amzn_fractals.iterrows():
    if row['fractals_high'] == True:
        bearish_markers.append(row['High']*1.1)
    else:
        bearish_markers.append(np.nan)
```

```python
amzn_fractals["bullish_markers"] = bullish_markers
amzn_fractals["bearish_markers"] = bearish_markers
```
